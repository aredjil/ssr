## Noisy Space Reducing Process for Different Values of $\lambda$

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
class Ball:
    """
    Ball class 
    """
    def __init__(self, state:int):
        # random number generator 
        self.rng = np.random.default_rng() 
        self.init_state = state # Intial state  
        self.state = state # current state 
        pass
    def reset(self):
        # Reset the state of the ball
        self.state = self.init_state  
```

```python
class SSR_process:
    def __init__(self, init_state=20, max_iter=100_000):
        self.init_state = init_state
        self.n_states = init_state
        self.max_iter = max_iter 
        ball = Ball(state=init_state)
        self.balls = [ball]
        self.rng = np.random.default_rng()
    def run(self, mu=2):
        visits_dict = defaultdict(int)
        # iterate the list of balls 
        for b in self.balls:
            for i in range(self.max_iter):
                next_state = self.rng.choice(range(1, b.state))
                b.state = next_state 
                visits_dict[b.state] += 1
                if b.state == 1:
                    b.reset()
        return visits_dict
            
```

```python
p = SSR_process()
```

```python
visits = p.run(mu=1)
```

```python
visits = {k:v/sum(visits.values()) for k, v in visits.items()}
visits = dict(sorted(visits.items()))
plt.loglog(visits.keys(), visits.values())
plt.xlabel("State")
plt.ylabel("Frequency")
plt.show()
```
